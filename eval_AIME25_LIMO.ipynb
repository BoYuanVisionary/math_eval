```python
from vllm import LLM, SamplingParams
from transformers import AutoTokenizer
import re

# Initialize the model
llm = LLM(
    model="GAIR/LIMO",
    tensor_parallel_size=4,  # adjust based on available GPUs
    trust_remote_code=True,
    # swap_space=40, # default is 60 but we don't have enough space； danguaous command don't use it
    gpu_memory_utilization=0.9,
)

# Configure generation parameters
sampling_params = SamplingParams(
    temperature=0.7,
    max_tokens=32768,
    top_p=0.95,
)

tokenizer = AutoTokenizer.from_pretrained("GAIR/LIMO", trust_remote_code=True)



```

```python
import data
generator = data.data_generator('Maxwell-Jia/AIME_2024')  
correct = 0
for i in range(20):
    prompt, answer = generator.generate_answer_2()
    messages = [
    {"role": "system", "content": "Please reason step by step, and put your final answer within \\boxed{}."},
    {"role": "user", "content": prompt}
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )
    o = llm.generate(text, sampling_params)
    print(o[0].outputs[0].text)
    print(f'answer: {answer}')
    pattern = r"\\boxed\{(\d+)\}"
    matches = re.findall(pattern, o[0].outputs[0].text)
    if len(matches) == 0:
        print(f'no matches found')
    else:    
        if len(set(matches)) != 1:
            print(f'matches: {matches} not unique')
            print(f'current answer: {answer}')
        # remove the leading 0s
        matches[-1] = matches[-1].lstrip('0')
        if matches[-1] == str(answer):
            correct += 1
    print(f'correct: {correct}, in {i+1} questions')

print(f'correct: {correct} out of 20')
```
